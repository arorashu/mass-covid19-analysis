```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# dataset = np.genfromtxt('us-states.csv', ',')
dataset = pd.read_csv('us-states.csv')
datasetMA = pd.read_csv('ma-covid19-daily.csv')
```

```python
dataset
datasetMA
```

```python
massCasesData = dataset[dataset["state"] == "Massachusetts"]
# massDailyTests = datasetMA[datasetMA["totalTestResultsIncrease"]]
```

```python
datasetMA['date'] = pd.to_datetime(datasetMA['date'], format = '%Y%m%d')
datasetMASorted = datasetMA.sort_values('date').reset_index(drop = True)
datasetMASorted
```

## Cumulative Tests and Cases

```python
plt.plot(datasetMA['date'],datasetMA['total'], 'r')
plt.plot(datasetMASorted['date'], datasetMASorted['positive'], 'b')
# plt.plot(datasetMA['date'],datasetMA['positive'] / datasetMA['totalTestResultsIncrease'])

plt.legend()
plt.plot()
plt.show()
```

## Daily Tests and Cases

```python
plt.plot(datasetMA['date'],datasetMA['totalTestResultsIncrease'], 'r')
plt.plot(datasetMASorted['date'], datasetMASorted['positiveIncrease'], 'b')
# plt.plot(datasetMA['date'],datasetMA['positive'] / datasetMA['totalTestResultsIncrease'])

plt.legend()
plt.plot()
plt.show()
```

```python
datasetMADropped = datasetMA.fillna(0)
plt.plot(datasetMADropped['date'],datasetMADropped['positiveIncrease'] / (datasetMADropped['totalTestResultsIncrease']+1) )

plt.plot()
plt.show()
```
